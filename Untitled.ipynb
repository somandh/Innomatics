```python
fight_club_ratings = movie_data_tags_links[movie_data_tags_links['title'] == 'Fight Club (1999)']

plt.figure(figsize=(8, 6))
plt.hist(fight_club_ratings['rating'], bins=10, color='skyblue')
plt.title('User Ratings Distribution for Fight Club (1999)')
plt.xlabel('User Ratings')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
desired=[858,318,2,106782]
hi=movie_data_tags_links[movie_data_tags_links['movieId'].isin(desired)]
hi
```

```python
average_ratings = hi.groupby('movieId')['rating'].mean().reset_index()

most_popular_movie = average_ratings.loc[average_ratings['rating'].idxmax()]
most_popular_movie
```

```python
user_ratings_count = movie_data_tags_links.groupby('title')['rating'].count()
user_ratings_count
```

```python
top_5_movies = movie_data_tags_links.sort_values(by='rating', ascending=False).head(5)

print(top_5_movies['title'].tolist())
```

```python
sci_fi_movies = ['Deadpool (2016)', 'Terminator 2: Judgment Day (1991)', 'X-Men: The Last Stand (2006)', 'Jurassic Park (1993)']
sci_fi_data = movie_data_tags_links[movie_data_tags_links['title'].isin(sci_fi_movies)]
user_ratings_count_sci_fi = sci_fi_data.groupby('title')['rating'].count().reset_index()

sorted_sci_fi = user_ratings_count_sci_fi.sort_values(by='rating', ascending=False)

third_most_popular_sci_fi = sorted_sci_fi.iloc[2]['title']

print(f"The third most popular Sci-Fi movie based on user ratings count is: {third_most_popular_sci_fi}")
```
